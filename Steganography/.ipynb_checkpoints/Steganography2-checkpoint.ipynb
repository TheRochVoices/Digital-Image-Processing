# Steganography
This is another code of steganography better than one I uploaded before.

```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = io.imread('lena.png')
```

```python
# copy of image just for visualization
img2=np.copy(img)
```

```python
# text to store in image
text="hey there"
```

```python
# plain text will be stored in binary form in this list
binary_text=list()
```

```python
# ord return ASCII code and bin convert ASCII code to binary
for i in text:
    binary_text.append(bin(ord(i)))
```

```python
# Removing 0b from the binary string
for i,j in enumerate(binary_text):
    binary_text[i]=j[2:]
```

```python
# if any number is not 7 bit, converting it in 7 bit
for i,j in enumerate(binary_text):
    if(len(binary_text[i])<7):
        binary_text[i]=binary_text[i]+'0'
```

```python
binary_text
```

```python
# a stream of binary number will be in to_cipher
to_cipher=str()
for i in binary_text:
    to_cipher+=i
```

```python
# converting stream of binary character into a list for simplicity
to_cipher=list(to_cipher)
```

```python
# last pixel of image will store the length of plain text
cipher_length=len(to_cipher)

img[-1][-1][-1]=cipher_length
```

```python
# Traversing in image and replacing Least Significant Bit of image with bit of our plain text. 
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(img.shape[2]):
            if(len(to_cipher)==0):
                break
            binary_pixel=bin(img[i][j][k])
            binary_pixel=binary_pixel[:-1]
            binary_pixel=binary_pixel+str(to_cipher.pop())
            img[i][j][k]=int(binary_pixel,2)
```

```python
plt.subplot(1,2,1)
plt.imshow(img2)
plt.xlabel('Original')
plt.subplot(1,2,2)
plt.xlabel('Ciphered')
plt.imshow(img)
```
