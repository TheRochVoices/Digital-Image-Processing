# Steganography
<p>Steganography is data hidden within data. Steganography is an encryption technique that can be used along with cryptography as an extra-secure method in which to protect data.</p>
<p>This notebook describes a way to hide message into an image</p>

```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
```

<p>Here img is the original image and img2 is image with hidden message</p> 

```python
img = io.imread('lena.png')
```

```python
img.shape
```

```python
img2=np.copy(img)
```

```python
text="Hey There"
```

```python
# Replacing the content of first row(red channel) with ASCII code of plain text(text variable)
for i,j in enumerate(text):
    img2[0,i,0]=ord(j)
```

```python
plt.subplot(1,2,1)
plt.xlabel('Normal')
plt.imshow(img)

plt.subplot(1,2,2)
plt.xlabel('Ciphered')
plt.imshow(img2)
```

# Decrypt
<p>If you subtract ciphered image with original image all pixel will become 0 except the ones replaced with ASCII code.</p>

```python
img3=img-img2
```

```python
recovery=list()
original=list()
```

```python
for i in range(img3.shape[0]):
    if(img3[0,i,0] ==0):
        break
    recovery.append(img3[0,i,0])
    original.append(img[0,i,0])
```

```python
print(recovery)
print(original)
```

<p>asci array contain the ASCII codes which are printed below.</p>

```python
asci=np.array(original)-np.array(recovery)
```

```python
asci
```

```python
for i in asci:
    print(chr(i))
```
