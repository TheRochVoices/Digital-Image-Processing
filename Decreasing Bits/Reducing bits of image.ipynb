# Decreasing number of bits of image

A greyscale image is represented by 8 bits(range from 0-255), in this program we are decreasing the number of bits in greyscale image by 1 and observing the loss in information.
<p>e.g. if we convert 8 bit image to 7 bit the new range will be 0-127 definately there will be loss in information.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io,color
```

```python
def to_grayscale(img):
    R=img[:,:,0]
    G=img[:,:,1]
    B=img[:,:,2]
    gray=((0.3 * R) + (0.59 * G) + (0.11 * B))
    return gray
```

```python
def reduce(img):
    return(img/2)
```

```python
img = io.imread('lena.png')
img=to_grayscale(img)
```

```python
image1=reduce(img).astype(np.int32)        #to 7 bit
image2=reduce(image1).astype(np.uint8)     #to 6 bit
image3=reduce(image2).astype(np.int32)     #to 5 bit
image4=reduce(image3).astype(np.int32)     #to 4 bit
image5=reduce(image4).astype(np.int32)     #to 3 bit
image5=reduce(image5).astype(np.int32)     #to 2 bit
image6=reduce(image5).astype(np.int32)     #to 1 bit
```

```python
plt.subplot(2, 3, 1)
plt.axis('off')
plt.imshow(image1,cmap='gray')

plt.subplot(2, 3, 2)
plt.axis('off')
plt.imshow(image2,cmap='gray')

plt.subplot(2, 3, 3)
plt.axis('off')
plt.imshow(image3,cmap='gray')

plt.subplot(2, 3, 4)
plt.axis('off')
plt.imshow(image4,cmap='gray')

plt.subplot(2, 3, 5)
plt.axis('off')
plt.imshow(image5,cmap='gray')

plt.subplot(2, 3, 6)
plt.axis('off')
plt.imshow(image6,cmap='gray')

plt.show()
```
